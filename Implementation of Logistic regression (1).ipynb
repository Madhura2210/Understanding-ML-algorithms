Logistic regression from scratch

```python
class Logistic_Regression:
            
    def split_X_y(self,df):
        df = df.drop_duplicates()
        return df.iloc[:,-8:-1].to_numpy(),df.iloc[:,-1].to_numpy().reshape(-1,1)
    
    def split_df_train_test_val(self,df):
        df_train = df.sample(frac=0.7) 
        values_to_exclude = df_train['Passengerid']
        mask_exclude = df['Passengerid'].isin(values_to_exclude)
        df_val =df[~mask_exclude].sample(frac=0.333333)
        values_to_exclude_2 = df_val['Passengerid']
        mask_exclude = df['Passengerid'].isin(pd.concat([values_to_exclude, values_to_exclude_2]))
        df_test =df[~mask_exclude]
        #split randomly without using random with seed values from scikit learn
        X_train,y_train = self.split_X_y(df_train)
        X_val,y_val = self.split_X_y(df_val)
        X_test,y_test = self.split_X_y(df_test)
        return X_train,y_train,X_val, y_val,X_test,y_test

        
    def partial_derivatives(self,X_train,y_train,y_pred):
        error = y_pred-y_train
        m = X_train.shape[0]
        dJdwa = np.dot(X_train.T,error)/m #916,7 and error 916,1, .t is needed
        dJdb = np.sum(error)/m
        return dJdwa ,dJdb
    
    def gradient_descent(self,X_train,y_train,iterations,alpha):
        # logit
        w = np.zeros((7,1))
        b = 0
        for i in range (iterations):
            y_pred = 1/(1+np.exp(-(np.dot(X_train,w)+b)))#916,7 and 7,1
            dJdwa, dJdb= self.partial_derivatives(X_train,y_train,y_pred)
            w = w - alpha*dJdwa
            b = b - alpha*dJdb
        return w,b
    
    def checkAccuracy(self,y,y_prob):
        y_pred = np.zeros((y_prob.shape[0],1))
        tp,fp,fn,tn=0,0,0,0
        bcel = 0
        for i in range(y_prob.shape[0]):
            if (y_prob[i]<0.5):
                y_pred[i]=0
            else:
                y_pred[i]=1
            if(y_pred[i]==1 and y[i]==1):
                tp = tp + 1
            elif (y_pred[i]==1 and y[i]==0):
                fp = fp + 1
            elif (y_pred[i]==0 and y[i]==1):
                fn = fn + 1
            elif (y_pred[i]==0 and y[i]==0):
                tn = tn + 1
        eps = 1e-15 # can not allow log 0
        y_pred_adjusted = np.clip(y_prob,eps,1-eps) #done with y_prob, not with y_pred
        bcel =-np.sum( y*np.log(y_pred_adjusted)+(1-y)*np.log(1-y_pred_adjusted))/(y.shape[0])
        mse = np.sum((y - y_pred)**2)
        
        
        return mse,tp,fp,fn,tn,bcel

obj = Logistic_Regression()   

X_train,y_train,X_val,y_val,X_test,y_test = obj.split_df_train_test_val(df)   

final_w,final_b = obj.gradient_descent(X_train,y_train,2000,0.01)

y_prob_final = 1/(1+np.exp(-(np.dot(X_train,final_w)+final_b)))
mse_train, tp_train,fp_train,fn_train,tn_train, bcel_train = obj.checkAccuracy(y_train,y_prob_final)

y_prob_val = 1/(1+np.exp(-(np.dot(X_val,final_w)+final_b)))
mse_val, tp_val,fp_val,fn_val,tn_val,bcel_val=obj.checkAccuracy(y_val, y_prob_val)  #can augement training with validation sets

y_prob_test = 1/(1+np.exp(-(np.dot(X_test,final_w)+final_b)))
mse_test, tp_test,fp_test,fn_test,tn_test,bcel_test=obj.checkAccuracy(y_test, y_prob_test )

print("We will print the total number of values that were misclassified. We will then proceed with printing confusion matrix")
print("Values misclassified in training set",mse_train)
print("TP",tp_train,"\tFP",fp_train,"\nFN",fn_train,"\tTN",tn_train)
print("Binary cross entropy loss in training ",bcel_train)

print("Values misclassified in validation set",mse_val)
print("TP",tp_val,"\tFP",fp_val,"\nFN",fn_val,"\tTN",tn_val)
print("Binary cross entropy loss in validation ",bcel_val)

print("Values misclassified in test set",mse_test)
print("TP",tp_test,"\tFP",fp_test,"\nFN",fn_test,"\tTN",tn_test)
print("Binary cross entropy loss in test ",bcel_test)
```

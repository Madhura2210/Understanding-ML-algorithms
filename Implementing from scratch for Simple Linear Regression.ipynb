```python
def partial_derivatives(X,y,y_pred):
    dJda= -np.inner(X,(y_pred-y))/len(y)
    dJdb= -np.sum(y_pred-y)/len(y)
    return dJda,dJdb
```

```python
def mod_weights(X,y,a,b,alpha):
    X= np.asarray(X).reshape(-1)
    y = np.asarray(y).reshape(-1)
    a = float(a)
    b = float(b)
    y_pred = a*X + b
    for i in range (4000):
        
        dJda, dJdb = partial_derivatives(X,y,y_pred)
        a = a + alpha*dJda
        b = b + alpha*dJdb
        y_pred = a*X+ b
        print("Mean Squared error for epoch ",i, "is ", np.inner((y_pred-y),(y_pred-y))/(30))
        
    return a,b
```

```python
class OLS:
    def calc_coefficient(X,y):
        numerator = 0.0
        denominator = 0.0
        for i in range (30):
            numerator = numerator + (X.iloc[i]-X_mean)*(y[i]-y_mean)
            denominator = denominator + (X.iloc[i]-X_mean)**2
            
        intercept = y_mean - numerator*X_mean/denominator
        slope = numerator/denominator
        return slope,intercept
    slope , intercept = calc_coefficient(X,y)
    print ("Slope is: ",slope, " Intercept is: ", intercept)
```
